## 4.2 price

```python
ds['price'].describe()
```

```python
# create Histogram

plt.figure(figsize=(10, 6))
sns.histplot(ds['price'],bins=20, kde=True)

# set title and label
plt.xlabel('price')
plt.ylabel('Frequency')
plt.title('Distribution of price Column')

# Display the plot
plt.show()
```

### Let's  apply logrithmic trasformation for better understanding of data distribution

```python
#apply logrithmic trasformation

log_prices = np.log(ds['price'])

# create a histogram

plt.figure(figsize=(10, 6))
sns.histplot(log_prices, bins=20, kde=True)

# set labels and title
plt.xlabel('Logarithm of Price')
plt.ylabel('Frequency')
plt.title('Distribution of Logarithm of price Column')

#Display the plot
plt.show()
```
